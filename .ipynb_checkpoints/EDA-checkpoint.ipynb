```python
train = pd.read_csv('pps_train.csv', index_col = "id")
test = pd.read_csv('pps_test.csv', index_col = "id")
```

```python
for col in train.columns:
    print(col, " : ", train[col].unique())
```

```python
train.cov()
```

```python
target = train['nerdy']
X = train.drop(['nerdy'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.2, shuffle=True, random_state=42, stratify=target)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(data = X_train_scaled, index = X_train.index, columns = X_train.columns)
X_test_scaled = pd.DataFrame(data = X_test_scaled, index = X_test.index, columns = X_test.columns)
```
